```python
import numpy as np

#numero de corredores e voltas dos corredores

num_corredores = 6
num_voltas = 10

#lista para armazenar nomes dos corredores

nomes_corredores = []

#dicionário para armazenar os tempos de cada volta de cada corredor

tempos_corredores = {}

# solicitar nomes dos corredores

for i in range(num_corredores):
  nome = input(f"Digite o nome do corredor {i+1}:")
  nomes_corredores.append(nome)
  tempos_corredores[nome] = []

# Solicitar tempo de cada volta de cada corredor
for nome_corredor in nomes_corredores:
    print(f"\nTempos do corredor {nome_corredor}:")
    for volta in range(num_voltas):
        tempo = float(input(f"Volta {volta+1} (em segundos): "))
        tempos_corredores[nome_corredor].append(tempo)

# Encontrar a melhor volta da prova
melhor_volta = np.min([np.min(tempos) for tempos in tempos_corredores.values()])
corredor_melhor_volta = [corredor for corredor, tempos in tempos_corredores.items() if melhor_volta in tempos][0]

print(f"A melhor volta da prova foi do corredor {corredor_melhor_volta}.")

# Calcular a classificação final em ordem crescente
classificacao_final = [(nome_corredor, np.sum(tempos)) for nome_corredor, tempos in tempos_corredores.items()]
classificacao_final.sort(key=lambda x: x[1])

print("\nClassificação final em ordem crescente:")
for posicao, (corredor, tempo_total) in enumerate(classificacao_final, start=1):
    print(f"{posicao}º lugar: {corredor} - Tempo total: {tempo_total:.2f} segundos")

# Encontrar a volta com a média mais rápida
media_por_volta = np.mean(list(tempos_corredores.values()), axis=0)
volta_media_rapida = np.argmin(media_por_volta) + 1

print(f"\nA volta com a média mais rápida foi a volta {volta_media_rapida}.")
```
